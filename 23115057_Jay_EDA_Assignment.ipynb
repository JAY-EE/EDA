Fills the missing values with "Unkown" in the columns with object type i.e columns except numerical data.

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

Fills the missing values with mean in the columns with numerical type i.e columns with numerical data.

```python
numerical_cols = df.select_dtypes(include=['number']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
```

Box plot of Luxembourg country based on the activity quantile.

Graph of top 10 and bottom 10 countries based on activity percentage.

```python

top_countries = df.groupby("country")["activity_percentage"].mean().nlargest(10)
bottom_countries = df.groupby("country")["activity_percentage"].mean().nsmallest(10)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.barplot(x=top_countries.values, y=top_countries.index, ax=axes[0], hue=top_countries.index, palette="Blues_r")
axes[0].set_title("Top 10 Countries by Activity Percentage")

sns.barplot(x=bottom_countries.values, y=bottom_countries.index, ax=axes[1], hue=bottom_countries.index, palette="Reds_r")
axes[1].set_title("Bottom 10 Countries by Activity Percentage")

plt.tight_layout()
plt.show()
```
